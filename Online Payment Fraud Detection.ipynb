```python
type_new = pd.get_dummies(data['type'], drop_first=True)
type_new
```

```python
data_new = pd.concat([data, type_new], axis=1)
data_new.head()
```

```python
y = data_new['isFraud']
y
```

```python
X = data_new.drop(['isFraud', 'type', 'nameOrig', 'nameDest'], axis=1)
X
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
```

```python
train_preds = lr_model.predict_proba(X_train)[:, 1]
print(f'{lr_model} : ')
print('Training Accuracy (ROC AUC) : ', roc_auc_score(y_train, train_preds))

y_preds = lr_model.predict_proba(X_test)[:, 1]
print('Validation Accuracy (ROC AUC) : ', roc_auc_score(y_test, y_preds))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
train_preds = lr_model.predict(X_train)
train_probs = np.clip(train_preds, 0, 1)
print(f'{lr_model} : ')
print('Training Accuracy (ROC AUC) : ', roc_auc_score(y_train, train_probs))

test_preds = lr_model.predict(X_test)
test_probs = np.clip(test_preds, 0, 1)
print('Validation Accuracy (ROC AUC) : ', roc_auc_score(y_test, test_probs))
```

```python
rf_model = RandomForestClassifier(n_estimators=7, criterion='entropy', random_state=7)
rf_model.fit(X_train, y_train)
```

```python
train_preds = rf_model.predict_proba(X_train)[:, 1]
print(f'{rf_model} : ')
print('Training Accuracy (ROC AUC) : ', roc_auc_score(y_train, train_preds))

y_preds = rf_model.predict_proba(X_test)[:, 1]
print('Validation Accuracy (ROC AUC) : ', roc_auc_score(y_test, y_preds))
```

```python
ConfusionMatrixDisplay.from_estimator(rf_model, X_test, y_test)
plt.title('Confusion Matrix for RandomForestClassifier')
plt.show()
```
